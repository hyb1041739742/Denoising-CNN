```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
path = r"C:\Users\bbfss\Desktop\file\Code\github\Denoising-CNN\Denoising-CNN_2.0\实验数据保存\1"
Real = list(np.loadtxt("real_A100000.txt"))
fake1 = list(np.loadtxt("fake_A2B100000.txt"))
fake2 = list(np.loadtxt("fake_B2A2B100000.txt"))
```

```python
def save_png(title, signal):
    fig = plt.figure(figsize=(30, 10))
    ax = fig.add_subplot(1,1,1)
    plt.xlim(0,2001)
    plt.ylim(-1,1)
    plt.yticks(np.arange(-1, 1.1, 0.4), size=40)
    plt.xticks(range(0, 2001, 400), size=40) # 设置网格跨度为0到100， 间距为20
    ax.plot(signal)
    plt.savefig(os.path.join(path, "{}.png".format(title)), dpi=600)
```

```python
save_png("Real", Real)
save_png("fake1", fake1)
save_png("fake2", fake2)
```
