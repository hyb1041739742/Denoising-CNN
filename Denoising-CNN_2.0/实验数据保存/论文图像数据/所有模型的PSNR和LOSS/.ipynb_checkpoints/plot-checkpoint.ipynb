```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
DUDnCNN_PSNR = list(np.loadtxt("DUDnCNN_PSNR.txt"))
FDUDnCNN_PSNR = list(np.loadtxt("FDUDnCNN_PSNR.txt"))
FDUDDnCNN_Improve_PSNR = list(np.loadtxt("FDUDDnCNN_Improve_PSNR.txt"))
FDUDDnCNN_PSNR = list(np.loadtxt("FDUDDnCNN_PSNR.txt"))
```

### 画PSNR的大图

```python
plt.figure(figsize=(8, 5))
plt.xlim(0, 200)
plt.ylim(25, 50)
plt.grid(axis="x", linestyle="-.", linewidth=1)
plt.grid(axis="y", linestyle="-.", linewidth=1)
plt.xticks(range(0, 201, 40), size=20) # 设置网格跨度为0到100， 间距为20
plt.yticks(range(25, 51, 5), size=20)
plt.tick_params(
    direction="out",  # 刻度线朝内
    length=10,
)
plt.plot(
    range(1, 201),
    DUDnCNN_PSNR[0:200],
    color="#06f906",
    label="DUDnCNN",
    marker="o",
    markerfacecolor="white",
    markeredgewidth=1.5,
    markersize=5.5,
    markevery=7,
)
plt.plot(
    range(1, 201),
    FDUDnCNN_PSNR[0:200],
    color="#4770eb",
    label="FDUDnCNN",
    marker="s",
    markerfacecolor="white",
    markeredgewidth=1.5,
    markersize=5.5,
    markevery=7,
)

plt.plot(
    range(1, 201),
    FDUDnCNN_Improve_PSNR[0:200],
    color="#45c9e0",
    label="FDUDnCNN_Improve",
    marker="^",
    markerfacecolor="white",
    markeredgewidth=1.5,
    markersize=5.5,
    markevery=7,
)

plt.plot(
    range(1, 201),
    FDUDDnCNN_PSNR[0:200],
    color="#cf3269",
    label="FDUDDnCNN",
    marker="D",
    markerfacecolor="white",
    markeredgewidth=1.5,
    markersize=5.5,
    markevery=7,                #每几格 标一个点
)

plt.legend(fontsize=13, loc="lower right")
plt.savefig("PSNR对比_大.png", dpi=600)
plt.show()
```
