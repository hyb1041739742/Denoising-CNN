```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
num = 8
Clean = list(np.loadtxt("Clean_Real_signal {}.txt".format(num)))
FDUDnCNN_Improve_Conv = list(np.loadtxt("FDUDnCNN_improve_Conv_Feature {}.txt".format(num)))
FDUDDnCNN_Conv = list(np.loadtxt("FDUDDnCNN_Conv_Feature {}.txt".format(num)))
FDUDnCNN_Improve = list(np.loadtxt("FDUDnCNN_improve_Real_signal {}.txt".format(num)))
FDUDDnCNN = list(np.loadtxt("FDUDDnCNN_Real_signal {}.txt".format(num)))
```

```python
def save_png(title, signal):
    fig = plt.figure(figsize=(30, 10))
    ax = fig.add_subplot(1,1,1)
    plt.xlim(0,6000)
    plt.ylim(-1,1)
    plt.yticks(np.arange(-1, 1.1, 0.4), size=40)
    plt.xticks(range(0, 6001, 1000), size=40) # 设置网格跨度为0到100， 间距为20
    ax.plot(signal)
    plt.savefig("{}.png".format(title), dpi=600)
```

```python
def save_feature(title, signal):
    fig = plt.figure(figsize=(30, 10))
    ax = fig.add_subplot(1,1,1)
    plt.xlim(0,6001)
    plt.ylim(0,20)
    plt.yticks(np.arange(0,21,5), size=40)
    plt.xticks(range(0, 6001, 1000), size=40) # 设置网格跨度为0到100， 间距为20
    ax.plot(signal)
    plt.savefig("{}.png".format(title), dpi=600)
```

```python
save_feature("FDUDnCNN_Improve_Conv {}".format(num), FDUDnCNN_Improve_Conv)
save_feature("FDUDDnCNN_Conv {}".format(num), FDUDDnCNN_Conv)
```

```python
save_png("Clean {}".format(num), Clean)
save_png("FDUDnCNN_Improve {}".format(num), FDUDnCNN_Improve)
save_png("FDUDDnCNN {}".format(num), FDUDDnCNN)
```
