```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Adam_PSNR = list(np.loadtxt("Adam_PSNR.txt"))
AdamW_PSNR = list(np.loadtxt("AdamW_PSNR.txt"))
RMSprop_PSNR = list(np.loadtxt("RMSprop_PSNR.txt"))
SGD_PSNR = list(np.loadtxt("SGD_PSNR.txt"))

Adam_loss = list(np.loadtxt("Adam_loss.txt"))
AdamW_loss = list(np.loadtxt("AdamW_loss.txt"))
RMSprop_loss = list(np.loadtxt("RMSprop_loss.txt"))
SGD_loss = list(np.loadtxt("SGD_loss.txt"))
```

### 优化器PSNR

```python
plt.figure(figsize=(8, 5))
plt.xlim(0, 200)
plt.ylim(25, 50)
plt.grid(axis="x", linestyle="-.", linewidth=1)
plt.grid(axis="y", linestyle="-.", linewidth=1)
plt.xticks(range(0, 201, 40), size=20) # 设置网格跨度为0到100， 间距为20
plt.yticks(range(25, 51, 5), size=20)
plt.tick_params(
    direction="out",  # 刻度线朝内
    length=10,
)
plt.plot(
    range(1, 201),
    Adam_PSNR[0:200],
    color="#06f906",
    label="Adam",
    marker="o",
    markerfacecolor="white",
    markeredgewidth=1.5,
    markersize=5.5,
    markevery=7,
)
plt.plot(
    range(1, 201),
    AdamW_PSNR[0:200],
    color="#4770eb",
    label="AdamW",
    marker="s",
    markerfacecolor="white",
    markeredgewidth=1.5,
    markersize=5.5,
    markevery=7,
)

plt.plot(
    range(1, 201),
    RMSprop_PSNR[0:200],
    color="#45c9e0",
    label="RMSprop",
    marker="^",
    markerfacecolor="white",
    markeredgewidth=1.5,
    markersize=5.5,
    markevery=7,
)

plt.plot(
    range(1, 201),
    SGD_PSNR[0:200],
    color="#cf3269",
    label="SGD",
    marker="D",
    markerfacecolor="white",
    markeredgewidth=1.5,
    markersize=5.5,
    markevery=7,                #每几格 标一个点
)

plt.legend(fontsize=13, loc="lower right")
plt.savefig("优化器_PSNR对比.png", dpi=600)
plt.show()
```
