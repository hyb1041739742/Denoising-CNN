```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
e3_PSNR = list(np.loadtxt("0.001_PSNR.txt"))
e2_PSNR = list(np.loadtxt("0.01_PSNR.txt"))
e4_PSNR = list(np.loadtxt("0.0001_PSNR.txt"))

e3_loss = list(np.loadtxt("0.001_loss.txt"))
e2_loss = list(np.loadtxt("0.01_loss.txt"))
e4_loss = list(np.loadtxt("0.0001_loss.txt"))
```

### 学习率PSNR

```python
plt.figure(figsize=(8, 5))
plt.xlim(0, 200)
plt.ylim(25, 50)
plt.grid(axis="x", linestyle="-.", linewidth=1)
plt.grid(axis="y", linestyle="-.", linewidth=1)
plt.xticks(range(0, 201, 40), size=20) # 设置网格跨度为0到100， 间距为20
plt.yticks(range(25, 51, 5), size=20)
plt.tick_params(
    direction="out",  # 刻度线朝内
    length=10,
)
plt.plot(
    range(1, 201),
    e2_PSNR[0:200],
    color="#06f906",
    label="Learning rate=0.01",
    marker="o",
    markerfacecolor="white",
    markeredgewidth=1.5,
    markersize=5.5,
    markevery=7,
)
plt.plot(
    range(1, 201),
    e3_PSNR[0:200],
    color="#4770eb",
    label="Learning rate=0.001",
    marker="s",
    markerfacecolor="white",
    markeredgewidth=1.5,
    markersize=5.5,
    markevery=7,
)

plt.plot(
    range(1, 201),
    e4_PSNR[0:200],
    color="#45c9e0",
    label="Learning rate=0.0001",
    marker="^",
    markerfacecolor="white",
    markeredgewidth=1.5,
    markersize=5.5,
    markevery=7,
)

# plt.plot(
#     range(1, 201),
#     SGD_PSNR[0:200],
#     color="#cf3269",
#     label="SGD",
#     marker="D",
#     markerfacecolor="white",
#     markeredgewidth=1.5,
#     markersize=5.5,
#     markevery=7,                #每几格 标一个点
# )

plt.legend(fontsize=13, loc="lower right")
plt.savefig("学习率_PSNR对比.png", dpi=600)
plt.show()
```
