```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
Clean = list(np.loadtxt("Clean.txt"))
Noisy = list(np.loadtxt("Noisy.txt"))
FDUDnCNN_Improve_Conv3 = list(np.loadtxt("FDUDnCNN_Improve_Conv3.txt"))
FDUDDnCNN_Conv3 = list(np.loadtxt("FDUDDnCNN_Conv3.txt"))
FDUDnCNN_Improve = list(np.loadtxt("FDUDnCNN_Improve.txt"))
FDUDDnCNN = list(np.loadtxt("FDUDDnCNN.txt"))
```

```python
def save_png(title, signal):
    fig = plt.figure(figsize=(30, 10))
    ax = fig.add_subplot(1,1,1)
    plt.xlim(0,2001)
    plt.ylim(-1,1)
    plt.yticks(np.arange(-1, 1.1, 0.4), size=40)
    plt.xticks(range(0, 2001, 400), size=40) # 设置网格跨度为0到100， 间距为20
    ax.plot(signal)
    plt.savefig("{}.png".format(title), dpi=600)
```

```python
def save_feature(title, signal):
    fig = plt.figure(figsize=(30, 10))
    ax = fig.add_subplot(1,1,1)
    plt.xlim(0,2001)
    plt.ylim(0,12)
    plt.yticks(np.arange(0,13,3), size=40)
    plt.xticks(range(0, 2001, 400), size=40) # 设置网格跨度为0到100， 间距为20
    ax.plot(signal)
    plt.savefig("{}.png".format(title), dpi=600)
```

```python
save_feature("FDUDnCNN_Improve_Conv3", FDUDnCNN_Improve_Conv3)
save_feature("FDUDDnCNN_Conv3", FDUDDnCNN_Conv3)
```

```python
save_png("Clean", Clean)
save_png("Noisy", Noisy)
save_png("FDUDnCNN_Improve", FDUDnCNN_Improve)
save_png("FDUDDnCNN", FDUDDnCNN)
```
