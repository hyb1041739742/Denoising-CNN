```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
orignal_PSNR = list(np.loadtxt("orignal_PSNR.txt"))
L2_PSNR = list(np.loadtxt("L2_PSNR.txt"))


orignal_loss = list(np.loadtxt("orignal_loss.txt"))
L2_loss = list(np.loadtxt("L2_loss.txt"))
```

### 正则的使用

```python
plt.figure(figsize=(8, 5))
plt.xlim(0, 200)
plt.ylim(25, 50)
plt.grid(axis="x", linestyle="-.", linewidth=1)
plt.grid(axis="y", linestyle="-.", linewidth=1)
plt.xticks(range(0, 201, 40), size=20) # 设置网格跨度为0到100， 间距为20
plt.yticks(range(25, 51, 5), size=20)
plt.tick_params(
    direction="out",  # 刻度线朝内
    length=10,
)
plt.plot(
    range(1, 201),
    orignal_PSNR[0:200],
    color="#06f906",
    label="Orignal",
    marker="o",
    markerfacecolor="white",
    markeredgewidth=1.5,
    markersize=5.5,
    markevery=7,
)
plt.plot(
    range(1, 201),
    L2_PSNR[0:200],
    color="#4770eb",
    label="L2 Regularization",
    marker="s",
    markerfacecolor="white",
    markeredgewidth=1.5,
    markersize=5.5,
    markevery=7,
)

# plt.plot(
#     range(1, 201),
#     e4_PSNR[0:200],
#     color="#45c9e0",
#     label="Learning rate=0.0001",
#     marker="^",
#     markerfacecolor="white",
#     markeredgewidth=1.5,
#     markersize=5.5,
#     markevery=7,
# )

# plt.plot(
#     range(1, 201),
#     SGD_PSNR[0:200],
#     color="#cf3269",
#     label="SGD",
#     marker="D",
#     markerfacecolor="white",
#     markeredgewidth=1.5,
#     markersize=5.5,
#     markevery=7,                #每几格 标一个点
# )

plt.legend(fontsize=13, loc="lower right")
plt.savefig("使用正则化_PSNR对比.png", dpi=600)
plt.show()
```
